```
import pandas as pd 
import matplotlib.pyplot as plt
```

```
data= pd.read_csv('http://13.234.66.67/summer19/datasets/bank.csv')
```

```
data.head(5)
```

```
plt.bar(data.Age,data.CreditScore)
plt.xlabel('Age')
plt.ylabel('CreditScore')
```

```
plt.scatter(data.Age,data.EstimatedSalary)
plt.xlabel('Age')
plt.ylabel('EstimatedSalary')
```

```
plt.bar(data.Age,data.EstimatedSalary)
plt.xlabel('Age')
plt.ylabel('EstimatedSalary')
```

```
plt.stackplot(data.Age,data.Balance)
plt.xlabel('Age')
plt.ylabel('Balance')
```

```
plt.bar(data.Age,data.Balance)
plt.xlabel('Age')
plt.ylabel('Balance')
```

```
data.groupby('Gender').size().plot('pie',shadow=True, label='Gender' ,autopct='%1.1f%%')
```

```
data.groupby('Geography').size().plot('pie', shadow=True,label='Geography' ,autopct='%1.1f%%')
```

```
data.groupby('Geography').size().plot('bar')
```

```
plt.stackplot(data.Gender,data.EstimatedSalary)
plt.xlabel('Gender')
plt.ylabel('EstimatedSalary')
```

```
plt.bar(data.Gender,data.EstimatedSalary)
plt.xlabel('Gender')
plt.ylabel('EstimatedSalary')
```
