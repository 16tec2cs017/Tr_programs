```
import cv2
from google.colab.patches import cv2_imshow
```

```
image1=cv2.imread('Sunset1.jpg')
image2=cv2.imread('sunset.jpeg')
```

```
new_sub=image1-image2
```

```
cv2_imshow(new_sub)
```
