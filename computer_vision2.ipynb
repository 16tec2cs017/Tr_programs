```
import cv2
from google.colab.patches import cv2_imshow
```

```
image=cv2.imread('sunset.jpeg')
```

```
cv2_imshow(image)
```

```
scaleX = 0.9
scaleY = 0.9
scaleUp = cv2.resize(image, None, fx= scaleX*3, fy= scaleY*3, interpolation= cv2.INTER_LINEAR)
```

```
cv2_imshow(scaleUp)
```

```
scaleUp1 = cv2.resize(image[100:150,100:150], None, fx= scaleX*5, fy= scaleY*5, interpolation= cv2.INTER_LINEAR)
```

```
cv2_imshow(scaleUp1)
```
